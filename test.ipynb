```python
from langchain_huggingface import HuggingFacePipeline
from transformers import pipeline
from langchain.prompts import PromptTemplate
from langchain_core.prompt_values import StringPromptValue

```

```python
summarization_pipeline = pipeline(task = "summarization", model = "facebook/bart-large-cnn", device = 0)
summerizer = HuggingFacePipeline(pipeline = summarization_pipeline)
```

```python
refinement_pipeline = pipeline(task = "summarization", model = "facebook/bart-large", device = 0)
refiner = HuggingFacePipeline(pipeline = refinement_pipeline)
```

```python
qa_pipeline = pipeline(task = "question-answering", model = "deepset/roberta-base-squad2", device = 0 )
qa = HuggingFacePipeline(pipeline = qa_pipeline)
```

```python
summary_template = PromptTemplate.from_template("Summarize the following text in {length} length : {Text}")

summarization_chain = summary_template| summerizer | refiner
```

```python
Text = input("Enter text to be summarized : ")

length = input("preferred length of the summary (long/medium/short) : ")

# formatted_prompt = summary_template.invoke({"text" : text, "length" : length})

# initial_summary = summerizer.invoke(formatted_prompt)

# initial_summaryP = StringPromptValue(text = initial_summary)

# final_summary = refiner.invoke(initial_summaryP)
final_summary = summarization_chain.invoke({"Text" : Text, "length" : length})

print(final_summary)

question = input("You can ask me a Question!(Enter'EXIT' if you wish to exit !!!!)")

while True:
  if question == "EXIT":
    print("Well then, GoodBye!!!")
    break
  else:
    # Access the underlying Hugging Face pipeline using the .pipeline attribute
    qa_result = qa.pipeline({"question" : question, "context" : final_summary})
    print(qa_result["answer"])
    question = input("You can ask me a Question!(Enter'EXIT' if you wish to exit !!!!)")
    break
```
